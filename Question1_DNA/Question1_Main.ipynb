```python
# Library import 
import matplotlib.pyplot as plt
import logging

attempt = 1
inputpath = "Input_Data/Attempt"+str(attempt)+".txt"
outputpath = "Output_Data/OuputAttempt"+str(attempt)+".log"
plotpath = "Visualization/PlotAttempt"+str(attempt)+".png"

# Configure the root logger (or create a named logger)
logging.basicConfig(filename=outputpath, format='%(asctime)s - %(message)s', level=logging.INFO)

# Get a logger instance
logger = logging.getLogger()

logger.info("Path of the input file = "+inputpath)
# Data Import
DNAfile = open(inputpath)
DNAstrand = DNAfile.read()

logger.info("Input DNA Strand = " + DNAstrand)
logger.info("Total Length of DNA Strand = " + str(len(DNAstrand)))

# Analysis
nA = DNAstrand.count('A')
nC = DNAstrand.count('C')
nG = DNAstrand.count('G')
nT = DNAstrand.count('T')

logger.info("Count of A nucleotide = "+ str(nA))
logger.info("Count of C nucleotide = "+ str(nC))
logger.info("Count of G nucleotide = "+ str(nG))
logger.info("Count of T nucleotide = "+ str(nT))

# Data Plotting 
Nucleotides = ["A", "C", "G", "T"]
CountValue = [nA, nC, nG, nT]

vis = plt.bar(Nucleotides, CountValue, color = ['red', 'blue', 'black', 'green'])
plt.xlabel('Nucelotides')
plt.ylabel('No. of Nucleotides')
plt.bar_label(vis)
plt.savefig(plotpath)
logger.info("Outputting plot to path = "+plotpath)

```
