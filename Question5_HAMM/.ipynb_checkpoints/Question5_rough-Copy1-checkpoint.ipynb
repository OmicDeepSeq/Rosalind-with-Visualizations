```python
input = open("Input_Data/SampleData.txt")
input = input.read()
s = input.split('_')
del s[0]
for i in s:
    id = i[0:4]
    a = i.count('A')
    c = i.count('C')
    g = i.count('G')
    t = i.count('T')
    id = 'Rosalind_' + id
    ratio = 100*((c+g)/(a+c+g+t))
    ratio = str(ratio)
    print(id + ' ' + ratio)
```
